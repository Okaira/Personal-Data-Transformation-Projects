```python
import pandas as pd

file_path = "C:/Users/eguen/Downloads/electricity.csv"
df = pd.read_csv(file_path)

df.head()
```

```python
df['day'] = df['day'].str.replace("b'", "").str.replace("'", "").astype(int)
```

```python
df.head
```

```python
df['hour'] = (df['period'] * 24).round().astype(int)
df['minute'] = ((df['period'] * 24 * 60) % 60).round().astype(int)
df.head()
```

```python


df['date'] = pd.date_range(start='2020-01-01', periods=len(df), freq='D')
df['timestamp'] = df['date'] + pd.to_timedelta(df['hour'], unit='h')
```

```python
df.head()
```

```python
df['year'] = df['timestamp'].dt.year
df['month'] = df['timestamp'].dt.month
df['day_of_week'] = df['timestamp'].dt.dayofweek
df['is_weekend'] = df['day_of_week'].isin([5,6]).astype(int)
df['hour'] = df['timestamp'].dt.hour
df.head()
```

```python
df['class'] = df['class'].str.replace("b'", "").str.replace("'", "")
df['class_label'] = df['class'].map({'UP': 1, 'DOWN': 0})
```

```python
# NSW
df['nswdemand_lag_1'] = df['nswdemand'].shift(1)
df['nswdemand_lag_24'] = df['nswdemand'].shift(24)

df['nswprice_lag_1'] = df['nswprice'].shift(1)
df['nswprice_lag_24'] = df['nswprice'].shift(24)

# VIC
df['vicdemand_lag_1'] = df['vicdemand'].shift(1)
df['vicdemand_lag_24'] = df['vicdemand'].shift(24)

df['vicprice_lag_1'] = df['vicprice'].shift(1)
df['vicprice_lag_24'] = df['vicprice'].shift(24)
```

```python
df['nswdemand_roll_3'] = df['nswdemand'].rolling(3).mean()
df['nswdemand_roll_24'] = df['nswdemand'].rolling(24).mean()

df['vicdemand_roll_3'] = df['vicdemand'].rolling(3).mean()
df['vicdemand_roll_24'] = df['vicdemand'].rolling(24).mean()

df['nswprice_roll_3'] = df['nswprice'].rolling(3).mean()
df['nswprice_roll_24'] = df['nswprice'].rolling(24).mean()
```

```python
df['nsw_price_demand_ratio'] = df['nswprice'] / (df['nswdemand'] + 1e-6)
df['vic_price_demand_ratio'] = df['vicprice'] / (df['vicdemand'] + 1e-6)

df['price_spread'] = df['nswprice'] - df['vicprice']
df['demand_spread'] = df['nswdemand'] - df['vicdemand']
```

```python
import numpy as np

df['hour_sin'] = np.sin(2 * np.pi * df['hour'] / 24)
df['hour_cos'] = np.cos(2 * np.pi * df['hour'] / 24)

df['dow_sin'] = np.sin(2 * np.pi * df['day_of_week'] / 7)
df['dow_cos'] = np.cos(2 * np.pi * df['day_of_week'] / 7)
```

```python
df.head()
```

```python
df['transfer_abs'] = df['transfer'].abs()
df['transfer_direction'] = (df['transfer'] > 0).astype(int)
```

```python
df = df.dropna().reset_index(drop=True)
```
