```python
file_path = "/mnt/c/Users/eguen/Downloads/amazon.csv"
import pandas as pd
df = pd.read_csv(file_path)
print(df.head())
```

```python
print(df.head())
```

```python
df['discounted_price'] = df['discounted_price'].str.replace('₹','').str.replace(',','').astype(float)
df['actual_price'] = df['actual_price'].str.replace('₹','').str.replace(',','').astype(float)
```

```python
df['discount_percentage'] = df['discount_percentage'].str.replace('%','').astype(float)
```

```python
df['rating_count'] = (
    df['rating_count']
    .str.replace(',', '', regex=True)   # remove commas
    .fillna("0")                        # replace NaN with "0"
    .astype(int)                        # convert to integer
)
```

```python
df['popularity_score'] = df['rating'] * df['rating_count']
```

```python
df.head
```

```python
df[['main_category','sub_category','sub_sub_category','sub_sub_sub_category']] = (
    df['category'].str.split('|', n=3, expand=True)
)
```

```python
avg_ratings = df.groupby('main_category')['rating_count'].sum().reset_index()
print(avg_ratings)
```

```python
# Clean rating column
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Now group by main_category and compute average rating
avg_ratings = df.groupby('main_category')['rating'].mean().reset_index()
print(avg_ratings)
```

```python
df['discounted_price'] = (
    df['discounted_price']
    .astype(str)                               # force to string
    .str.replace('₹','', regex=True)           # remove currency symbol
    .str.replace(',','', regex=True)           # remove commas
    .str.replace('nan','', regex=True)         # handle NaN strings
    .replace('', '0')                          # replace empty with 0
    .astype(float)                             # convert to float
)

df['actual_price'] = (
    df['actual_price']
    .astype(str)
    .str.replace('₹','', regex=True)
    .str.replace(',','', regex=True)
    .str.replace('nan','', regex=True)
    .replace('', '0')
    .astype(float)
)
```
