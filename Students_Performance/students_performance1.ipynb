```python
# clean column names: lowercase and replace spaces with underscores
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
# strip whitespace and lowercase for consistency

cal_cols = ['gender', 'race/ethnicity', 'parental_level_of_education', 'lunch', 'test_preparation_course']

for col in cal_cols:
    df[col] = df[col].str.strip().str.lower()
```

```python
df.head()
```

```python
# average scores

df['average_score'] = df[['math_score', 'reading_score', 'writing_score']].mean(axis=1)
```

```python
print(df.columns.tolist())
```

```python
# score gap max-min

df['score_gap'] = df[['math_score', 'reading_score', 'writing_score']].max(axis=1) - \
                df[['math_score', 'reading_score', 'writing_score']].min(axis=1)
```

```python
df.head()
```

```python
# Prep Effectiveness Flag
# Flag students who completed the test preparation course and have an average score above 70 on average

df['prep_effectiveness'] = ((df['test_preparation_course'] == 'completed') & (df['average_score'] > 70)).astype(int)

```

```python
df.head()
```

```python
# Categorize performance

def categorize(score):
    if score >= 90:
        return 'excellent'
    elif score >= 75:
        return 'good'
    elif score >= 60:
        return 'average'
    else:
        return 'needs improvement'
    
df['performance_category'] = df['average_score'].apply(categorize)
```

```python
import pandas as pd
from sqlalchemy import create_engine

engine = create_engine("mysql+pymysql://root:opakaira@localhost/students_db")

query = """
SELECT gender,
       AVG(math_score) AS avg_math,
       AVG(reading_score) AS avg_reading,
       AVG(writing_score) AS avg_writing,
       AVG(average_score) AS overall_avg
FROM students_performance
GROUP BY gender;
"""
df_gender = pd.read_sql(query, engine)
print(df_gender)
```

```python
query_gender = """
SELECT gender,
       AVG(math_score) AS avg_math,
       AVG(reading_score) AS avg_reading,
       AVG(writing_score) AS avg_writing,
       AVG(average_score) AS overall_avg
FROM students_performance
GROUP BY gender;
"""
df_gender = pd.read_sql(query_gender, engine)
```

```python
query_parent = """
SELECT parental_level_of_education,
       SUM(CASE WHEN performance_category = 'High' THEN 1 ELSE 0 END) AS High,
       SUM(CASE WHEN performance_category = 'Medium' THEN 1 ELSE 0 END) AS Medium,
       SUM(CASE WHEN performance_category = 'Low' THEN 1 ELSE 0 END) AS Low
FROM students_performance
GROUP BY parental_level_of_education;
"""
df_parent = pd.read_sql(query_parent, engine)
```

```python
query_prep = """
SELECT test_preparation_course,
       AVG(average_score) AS avg_score
FROM students_performance
GROUP BY test_preparation_course;
"""
df_prep = pd.read_sql(query_prep, engine)
```

```python
query_race = """
SELECT race_ethnicity,
       AVG(average_score) AS avg_score
FROM students_performance
GROUP BY race_ethnicity;
"""
df_race = pd.read_sql(query_race, engine)
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# 1. Average scores by gender
df_gender.set_index('gender').plot(kind='bar', ax=axes[0,0])
axes[0,0].set_title("Average Scores by Gender")
axes[0,0].set_ylabel("Score")

# 2. Performance distribution by parental education
df_parent.set_index('parental_level_of_education').plot(kind='bar', stacked=True, ax=axes[0,1])
axes[0,1].set_title("Performance Distribution by Parental Education")
axes[0,1].set_ylabel("Number of Students")

# 3. Prep course effectiveness
df_prep.plot(x='test_preparation_course', y='avg_score', kind='bar', color=['red','green'], ax=axes[1,0])
axes[1,0].set_title("Prep Course Effectiveness")
axes[1,0].set_ylabel("Average Score")

# 4. Race/Ethnicity group analysis
df_race.plot(x='race_ethnicity', y='avg_score', kind='barh', color='skyblue', ax=axes[1,1])
axes[1,1].set_title("Average Score by Race/Ethnicity")
axes[1,1].set_xlabel("Average Score")

plt.tight_layout()
plt.show()
```
