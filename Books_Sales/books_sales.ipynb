```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
```

```python
# Standardize column names
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)

df.head(1)
```

```python
# Rename a couple of awkward ones for clarity
df = df.rename(columns={
    'publicationyear': 'publication_year',
    'pagecount': 'page_count',
    'most_popular_country': 'most_popular_country',
    'rating_average': 'rating_average'
})

# Strip strings
str_cols = df.select_dtypes(include='object').columns
for col in str_cols:
    df[col] = df[col].str.strip()

# Booleans: bestseller_status, adapted_to_movie
bool_cols = ['bestseller_status', 'adapted_to_movie']
for col in bool_cols:
    df[col] = df[col].map({'True': True, 'False': False}).astype('boolean')

# Numeric: movie_release_year, publication_year, page_count, rating_average
df['movie_release_year'] = pd.to_numeric(df['movie_release_year'], errors='coerce').astype('Int64')
df['publication_year'] = pd.to_numeric(df['publication_year'], errors='coerce').astype('Int64')
df['page_count'] = pd.to_numeric(df['page_count'], errors='coerce').astype('Int64')
df['rating_average'] = pd.to_numeric(df['rating_average'], errors='coerce')
```

```python
plt.figure(figsize=(8,4))
sns.histplot(df['rating_average'], bins=20, kde=True)
plt.title("Rating average distribution")
plt.show()
```

```python
dim_author = (
    df[['author']]
    .dropna()
    .drop_duplicates()
    .reset_index(drop=True)
    .reset_index(names='author_id')
)

dim_author['author_id'] += 1  # start ids at 1
dim_author.head()
```

```python
dim_genre = (
    df[['genre']]
    .dropna()
    .drop_duplicates()
    .reset_index(drop=True)
    .reset_index(names='genre_id')
)
dim_genre['genre_id'] += 1
dim_genre.head()
```

```python
dim_publisher = (
    df[['publisher']]
    .dropna()
    .drop_duplicates()
    .reset_index(drop=True)
    .reset_index(names='publisher_id')
)
dim_publisher['publisher_id'] += 1
dim_publisher.head()
```

```python
dim_country = (
    df[['most_popular_country']]
    .dropna()
    .drop_duplicates()
    .reset_index(drop=True)
    .reset_index(names='country_id')
)
dim_country['country_id'] += 1
dim_country.head()
```

```python
awards_series = (
    df['awards']
    .dropna()
    .astype(str)
    .str.split(',')
)

awards_exploded = (
    awards_series
    .explode()
    .str.strip()
    .replace('', np.nan)
    .dropna()
    .drop_duplicates()
    .reset_index(drop=True)
    .to_frame(name='award_name')
    .reset_index(names='award_id')
)
awards_exploded['award_id'] += 1

dim_award = awards_exploded
dim_award.head()
```

```python
dim_book = df[[
    'id',
    'title',
    'isbn',
    'page_count',
    'description',
    'language',
    'age_category',
    'tags',
    'publication_year'
]].drop_duplicates()

dim_book = dim_book.rename(columns={'id': 'book_id'})
dim_book.head()
```

```python
# Merge foreign keys into df copy
fact = df.copy()

fact = fact.merge(dim_author, on='author', how='left')
fact = fact.merge(dim_genre, on='genre', how='left')
fact = fact.merge(dim_publisher, on='publisher', how='left')
fact = fact.merge(dim_country, on='most_popular_country', how='left')

fact_book_popularity = fact[[
    'id',
    'author_id',
    'genre_id',
    'publisher_id',
    'country_id',
    'rating_average',
    'bestseller_status',
    'adapted_to_movie',
    'movie_release_year'
]].rename(columns={'id': 'book_id'})

fact_book_popularity.head()
```

```python
book_awards = (
    df[['id', 'awards']]
    .dropna()
    .rename(columns={'id': 'book_id'})
)

book_awards = (
    book_awards
    .assign(awards_list=lambda x: x['awards'].astype(str).str.split(','))
    .explode('awards_list')
)

book_awards['awards_list'] = book_awards['awards_list'].str.strip()
book_awards = book_awards[book_awards['awards_list'] != '']

book_awards = book_awards.merge(dim_award, left_on='awards_list', right_on='award_name', how='left')

bridge_book_award = book_awards[['book_id', 'award_id']].drop_duplicates()
bridge_book_award.head()
```

```python
dim_author.to_csv("dim_author.csv", index=False)
dim_genre.to_csv("dim_genre.csv", index=False)
dim_publisher.to_csv("dim_publisher.csv", index=False)
dim_country.to_csv("dim_country.csv", index=False)
dim_award.to_csv("dim_award.csv", index=False)
dim_book.to_csv("dim_book.csv", index=False)
fact_book_popularity.to_csv("fact_book_popularity.csv", index=False)
bridge_book_award.to_csv("bridge_book_award.csv", index=False)
```
