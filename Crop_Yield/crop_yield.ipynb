```python


# Standardize column names (snake_case)
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)

# Extract dimension tables
dim_soil_type = pd.DataFrame({"soil_type": df["soil_type"].unique()})
dim_region = pd.DataFrame({"region": df["region"].unique()})
dim_season = pd.DataFrame({"season": df["season"].unique()})
dim_crop_type = pd.DataFrame({"crop_type": df["crop_type"].unique()})
dim_irrigation_type = pd.DataFrame({"irrigation_type": df["irrigation_type"].unique()})

# Create surrogate keys
for dim in [dim_soil_type, dim_region, dim_season, dim_crop_type, dim_irrigation_type]:
    dim.insert(0, dim.columns[0] + "_id", range(1, len(dim) + 1))

# Merge back to create fact table
df_fact = (
    df
    .merge(dim_soil_type, on="soil_type")
    .merge(dim_region, on="region")
    .merge(dim_season, on="season")
    .merge(dim_crop_type, on="crop_type")
    .merge(dim_irrigation_type, on="irrigation_type")
)

# Drop original categorical columns
df_fact = df_fact.drop(columns=["soil_type", "region", "season", "crop_type", "irrigation_type"])
```

```python
df_fact.to_csv("C:/Users/eguen/Downloads/crop_yield_fact.csv", index=False)
```

```python

import seaborn as sns
import matplotlib.pyplot as plt

# Load your dataset
df = pd.read_csv("C:/Users/eguen/Downloads/crop_yield_fact.csv")

# -----------------------------
# 1. CORRELATION HEATMAP
# -----------------------------
plt.figure(figsize=(14, 10))
sns.heatmap(df.corr(numeric_only=True), annot=False, cmap="viridis")
plt.title("Correlation Heatmap")
plt.show()

# -----------------------------
# 2. DISTRIBUTION PLOTS
# -----------------------------
numeric_cols = [
    "soil_ph", "soil_moisture", "rainfall",
    "fertilizer_used", "pesticide_used",
    "crop_yield_ton_per_hectare"
]

for col in numeric_cols:
    plt.figure(figsize=(8, 5))
    sns.histplot(df[col], kde=True, color="teal")
    plt.title(f"Distribution of {col}")
    plt.show()

# -----------------------------
# 3. BOXPLOTS BY CATEGORY
# -----------------------------
category_cols = [
    "soil_type_id", "region_id", "season_id",
    "crop_type_id", "irrigation_type_id"
]

for col in category_cols:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=df[col], y=df["crop_yield_ton_per_hectare"])
    plt.title(f"Crop Yield by {col}")
    plt.show()

# -----------------------------
# 4. PAIRPLOT
# -----------------------------
pairplot_cols = [
    "soil_ph", "soil_moisture", "rainfall",
    "fertilizer_used", "pesticide_used",
    "crop_yield_ton_per_hectare"
]

```
